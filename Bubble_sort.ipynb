```python
# Time complexity is : O(n^2)
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:

                # swap the elements
                arr[j], arr[j+1] = arr[j+1], arr[j]

    return arr


arr = [70, 20, 50, 30, 90, 5, 15,22,17,60]

result = bubble_sort(arr)
print("Result of the sort is :", result)
```
